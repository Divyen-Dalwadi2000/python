# 1) Filling Null values

```python
df2 = df.fillna(value =0).head(10)   # it fill the null value with the 0 we can see in df(7,profit) =>> and asign to new dataframe called df2
```

```python
df2
```

```python
df2.isnull().sum().sum()
```

# filling null values whith previous values 

```python
df3 = df.fillna(method = 'pad')
df3
```

# filling null values whith Next values 

```python
df4 = df.fillna(method = 'bfill')
df4
```

# fill The null value column wise - previous value
      column1 => null         => column3
      column1 => column1 => column3


```python
df5 = df.fillna(method = 'pad' , axis = 1)
df5
```

# fill The null value column wise - comming / next value 
    column1 => null    => column3
    column1 => column3 => column3

```python
df6 = df.fillna(method = 'bfill' , axis = 1)
df6
```

```python
# filling different values in Null in different columns
df7 = df.fillna({'Region' : 'abc' ,
                'Discount': '_val_',})
df7
```

```python
# filling null value with the "mean" of a column
# we can use 'max' , 'min' .... insted of mean
df8 = df.fillna({'Discount' : df['Discount'].min()})
df8
```

# dropna()
     Drop the row(Which contain null value)

```python
df10 = df.dropna()
df10
```

```python
df10.isnull().sum().sum()
```

```python
# if there is in row any shell is null then "any" remove that row
df11 = df.dropna(how='any')
df11
```

```python
# if there is in row all shellis null then "all" remove that row
df12 = df.dropna(how='all')
df12
```

```python
# "numpy" is required for "replace" 
# here it replace specific null shell whith the given value
import numpy as np
df13 = df.replace(to_replace=np.nan , value = 123)
df13
```

```python
# Replace value with another value 
#                   original value , replace value
df15 = df.replace(to_replace=0.15 , value=555.0)
df15
```
